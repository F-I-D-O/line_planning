```python
import h5py
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
dm_files = [
    Path("./manhattan_dist_1.txt"),
    Path("./manhattan_dist_2.txt"),
    Path("./manhattan_dist_3.txt")
]

matrices = []
for dm_file in dm_files:
    matrices.append(np.loadtxt(dm_file))
```

```python
final_matrix = matrices[0] + matrices[1] + matrices[2]
```

```python
# fix too large values

# detect rows with too large values
threshold = 1e6
mask = (final_matrix > threshold).sum(axis=1) > 1000

# copy matrix
final_fixed_rows = final_matrix.copy()

# fix rows detected by the mask
last_valid = None
for i in range(len(final_matrix)):
    if not mask[i]:
        last_valid = final_matrix[i]
    elif last_valid is not None:
        final_fixed_rows[i] = last_valid

# now do the same for columns
mask = (final_matrix > threshold).sum(axis=0) > 1000
last_valid = None
for i in range(len(final_matrix)):
    if not mask[i]:
        last_valid = final_fixed_rows[:,i]
    elif last_valid is not None:
        final_fixed_rows[:,i] = last_valid
```

```python
with h5py.File('./dm.h5', 'w') as f:
    f.create_dataset('dm', data=final_fixed_rows)
```
